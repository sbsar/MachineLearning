```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
```

```python
cancer
```

```python
output = pd.DataFrame(cancer['target'])

dfs = [df,output]
```

```python
result = pd.concat(dfs, axis=1, join='inner')
```

```python

result.rename(columns={0:'Target'}, inplace= True)
```

```python
x = result.drop(['Target'], axis=1)
y = result['Target']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size=0.25, random_state=22)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(kernel='poly', degree=2, gamma=0.01, coef0=1, C=2)
```

```python
model.fit(x_train,y_train)
```

```python
pre = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, pre))
```

```python
print(confusion_matrix(y_test, pre))
```

c and gamma, gridsearch

```python
grid_param = {'C': [0.1,1, 10, 100, 1000],'gamma':[1,0.1,0.01,0.001,0.0001],'kernel': ['rbf'] }
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(estimator=SVC(), param_grid=grid_param, verbose=3)
```

```python
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```
