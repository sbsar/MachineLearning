```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
data = pd.DataFrame(pd.read_csv("data_input_output.csv"))
```

```python
x = data['input'].values.reshape(-1,1)
```

```python
y = data['output'].values.reshape(-1,1)
```

```python
plt.scatter(x,y, color="blue")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state = 20)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial_model_structure = PolynomialFeatures(degree = 5, include_bias = False)
```

```python
x_train_poly = polynomial_model_structure.fit_transform(x_train)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lrm = LinearRegression()
```

```python
lrm.fit(x_train_poly,y_train)
```

```python
x_test_poly = polynomial_model_structure.fit_transform(x_test)
```

```python
lrm.score(x_train_poly,y_train)
```

```python
lrm.score(x_test_poly,y_test)
```

```python
output = lrm.predict(x_test_poly)
```

```python
plt.plot(x_test, output)
plt.scatter(x_test, output, color="red", )
```
